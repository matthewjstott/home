```python
import os

def get_all_keys_in_directory(directory):
    file_paths = []
    
    # Walk through the directory and its subfolders
    for root, dirs, files in os.walk(directory):
        for file in files:
            # Join the root path with the file name to get the full file path
            file_paths.append(os.path.join(root, file))
    
    return file_paths

# Example usage:
directory_path = 'kioto/img/photography'  # Change this to the directory you want to scan
all_keys = get_all_keys_in_directory(directory_path)
all_keys = [key.replace('kioto/img','') for key in all_keys]


# Print all file paths
for file in all_keys:
    print(file)
```
